## Análise Exploratória de Dados (AED)

```python
import pandas as pd

# Leitura do arquivo CSV
df = pd.read_csv("C:/Users/karin/OneDrive/Área de Trabalho/EBAC/clientes-v2.csv")

# Visualização inicial
df.head()
```

### Verificação inicial

```python
# Visualização final
df.tail()
```

```python
# Conversão da coluna de data, se existir
df['data'] = pd.to_datetime(df['data'], format='%d/%m/%Y', errors='coerce')
```

```python
print('Verificação inicial:')
df.info()
```

###  Análise de dados nulos

```python
print('Análise de dados nulos:\n', df.isnull().sum())
print('% de dados nulos:\n', df.isnull().mean() * 100)
```

```python
# Remoção de linhas com valores nulos
df.dropna(inplace=True)
print('Confirmar remoção de dados nulos:\n', df.isnull().sum())
```

### Análise de dados duplicados

```python
print('Análise de dados duplicados:\n', df.duplicated().sum())
```

### Estatísticas e dados únicos

```python
print('Valores únicos por coluna:\n', df.nunique())
df.describe()
```

### Seleção de colunas e visualização

```python
df = df[['idade', 'data', 'estado', 'salario', 'nivel_educacao', 
         'numero_filhos', 'estado_civil', 'area_atuacao']]
df.head()
```

### Salvando o dataset tratado
