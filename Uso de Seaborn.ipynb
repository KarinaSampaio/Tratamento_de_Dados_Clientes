### Importações de bibliotecas

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### Leitura dos dados

```python
df = pd.read_csv("C:/Users/karin/OneDrive/Área de Trabalho/EBAC/clientes_preparados.csv")
df.head()
```

### Gráfico de Dispersão com sns.jointplot

```python
sns.jointplot(x='idade', y='salario', data=df, kind='scatter')
plt.show()
```

### Gráfico de Densidade (KDE)


```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df['salario'], fill=True, color='#863e9c')
plt.title('Densidade de Salários')
plt.xlabel('Salário')
plt.show()
```

### Gráfico de PairPlot – Dispersão e Histogramas


```python
sns.pairplot(df[['idade', 'salario', 'nivel_educacao']])
plt.show()
```

### Gráfico de Regressão

```python
sns.regplot(
    x='idade',
    y='salario',
    data=df,
    color='#27f865',
    scatter_kws={'alpha': 0.5, 'color': '#34c289'}
)
plt.title('Regressão de Salário por Idade')
plt.xlabel('Idade')
plt.ylabel('Salário')
plt.show()
```

### Gráfico countplot com hue

```python
sns.countplot(x='estado_civil', hue='nivel_educacao', data=df, palette='pastel')
plt.xlabel('Estado Civil')
plt.ylabel('Quantidade de Clientes')
plt.legend(title='Nível de Educação')
plt.show()
```
