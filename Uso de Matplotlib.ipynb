### Importação de bibliotecas

```python
import matplotlib.pyplot as plt
import pandas as pd
```

### Leitura do arquivo CSV

```python
df = pd.read_csv("C:/Users/karin/OneDrive/Área de Trabalho/EBAC/clientes_preparados.csv")
df.head(20)
```

### Gráfico de Barras – versão 1

```python
plt.figure(figsize=(10, 6))
df['nivel_educacao'].value_counts().plot(kind='bar', color='#90ee70')  # cor verde claro
plt.title('Divisão de Escolaridade - 1')
plt.xlabel('Nível de Educação')
plt.ylabel('Quantidade')
plt.xticks(rotation=0)
plt.show()
```

### Gráfico de Barras – versão 2 (manual com plt.bar)

```python
x = df['nivel_educacao'].value_counts().index
y = df['nivel_educacao'].value_counts().values

plt.figure(figsize=(10, 6))
plt.bar(x, y, color='#600a65')  # cor roxa escura
plt.title('Divisão de Escolaridade - 2')
plt.xlabel('Nível de Educação')
plt.ylabel('Quantidade')
plt.show()
```

###  Gráfico de Pizza (Pie Chart)

```python
plt.figure(figsize=(10, 6))
plt.pie(y, labels=x, autopct='%.1f%%', startangle=90)
plt.title('Distribuição de Nível de Educação')
plt.show()
```

### Gráfico de Dispersão com Hexbin (Idade vs Salário)

```python
plt.figure(figsize=(10, 6))

# Dispersão com hexbin detalhado
plt.hexbin(df['idade'], df['salario'], gridsize=40, cmap='Blues')
plt.colorbar(label='Contagem dentro do bin')  # Barra de cores ao lado
plt.xlabel('Idade')
plt.ylabel('Salário')
plt.title('Dispersão de Idade e Salário')

plt.show()

```
