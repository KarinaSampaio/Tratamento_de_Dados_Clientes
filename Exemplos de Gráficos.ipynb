### Importações

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### Leitura do arquivo

```python
df = pd.read_csv("C:/Users/karin/OneDrive/Área de Trabalho/EBAC/clientes_preparados.csv")
df.head()
```

### Heatmap de Correlação

```python
# Seleciona colunas numéricas relevantes para correlação
df_corr = df[['salario', 'idade', 'numero_filhos', 'nivel_educacao_ordinal', 
              'area_atuacao_cod', 'estado_cod']].corr()

# Plot do heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df_corr, annot=True, fmt=".2f", cmap="coolwarm")
plt.title('Mapa de Calor da Correlação entre Variáveis')
plt.show()
```

### Countplot simples (estado_civil)

```python
sns.countplot(x='estado_civil', data=df)
plt.title('Distribuição do Estado Civil')
plt.xlabel('Estado Civil')
plt.ylabel('Contagem')
plt.show()
```

###  Countplot com hue (por nivel_educacao)

```python
sns.countplot(x='estado_civil', hue='nivel_educacao', data=df)
plt.title('Distribuição do Estado Civil por Nível de Educação')
plt.xlabel('Estado Civil')
plt.ylabel('Contagem')
plt.legend(title='Nível de Educação')
plt.show()
```
