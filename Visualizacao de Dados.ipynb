###  Histograma simples

```python
# üìä Histograma simples
plt.hist(df['salario'])
plt.show()
```

### Histograma com par√¢metros personalizados

```python
plt.figure(figsize=(10, 6))
plt.hist(df['salario'], bins=100, color='green', alpha=0.8)
plt.title('Histograma - Distribui√ß√£o de Sal√°rios')
plt.xlabel('Sal√°rio')
plt.xticks(range(0, int(df['salario'].max()) + 2000, 2000))
plt.ylabel('Frequ√™ncia')
plt.grid(True)
plt.show()
```

### M√∫ltiplos gr√°ficos em uma figura

```python
plt.figure(figsize=(12, 6))
```

```python
# üîπ 1¬∫ Gr√°fico de Dispers√£o
plt.subplot(1, 2, 1)  # 1 linha, 2 colunas, 1¬∫ gr√°fico
plt.scatter(df['salario'], df['salario'])
plt.title('Dispers√£o - Sal√°rio x Sal√°rio')
plt.xlabel('Sal√°rio')
plt.ylabel('Sal√°rio')
```

```python
# üîπ 2¬∫ Gr√°fico de Dispers√£o
# 2¬∫ Gr√°fico de Dispers√£o com IDADE
plt.subplot(1, 2, 2)  # 1 linha, 2 colunas, 2¬∫ gr√°fico
plt.scatter(df['salario'], df['idade'], color='#5838a3', alpha=0.6, s=30)
plt.title('Dispers√£o - Sal√°rio x Idade')
plt.xlabel('Sal√°rio')
plt.ylabel('Idade')
```

### Mapa de calor de correla√ß√£o

```python
# üîç C√°lculo da correla√ß√£o
corr = df[['salario', 'idade']].corr()

# üî• Heatmap com Seaborn
plt.subplot(2, 2, 3)  # 2 linhas, 2 colunas, posi√ß√£o 3
sns.heatmap(corr, annot=True, cmap='Blues', vmin=-1, vmax=1)
plt.title('Mapa de Correla√ß√£o')

plt.tight_layout()
plt.show()
```
